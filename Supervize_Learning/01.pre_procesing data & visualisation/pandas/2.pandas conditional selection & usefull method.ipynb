```python
def yelp(price):
    if price<10:
        return '$'
    elif price>=10 and price<30:
        return '$$'
    else:
        return '$$$'
```

# apply method with lambda function

```python
#Ù…Ø­Ø§Ø³Ø¨Ù‡ Ù…Ø§Ù„ÛŒØ§Øª ØµÙˆØ±Øª Ø­Ø³Ø§Ø¨
df['total_bill'].apply(lambda bill : bill* 0.18)
```

```python
def quality(total_bill , tip):
    if tip/ total_bill >0.20:
        return 'Generuse'
    else:
        return 'other'
```

```python
# one way
df['Quality']=df[['total_bill','tip']].apply(lambda df :quality(df['total_bill'], df['tip']),axis=1)
```

```python
df.head(7)
```

```python
# other and faster way
df['Tip Quality']= np.vectorize(quality)(df['total_bill'],df['tip'])
```
