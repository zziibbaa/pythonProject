```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../Linear_Regression/Advertising.csv')
df
```

```python
X=df.drop('sales' , axis=1)
y=df['sales']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)

X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
model=Ridge(alpha=100)
```

# sklearn --> cross_val_score

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(model , X_train , y_train , cv=10 , scoring='neg_root_mean_squared_error')
score
```

```python
abs(score.mean())
```

```python
model.fit(X_train , y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error
MAE=mean_absolute_error(y_test , y_pred)
RMSE=np.sqrt(mean_squared_error(y_test , y_pred))
MAE , RMSE
```

```python
model=Ridge(alpha=1)
```

```python
score=cross_val_score(model , X_train , y_train , cv=10 , scoring='neg_root_mean_squared_error')
score
```

```python
abs(score.mean())
```

```python
model.fit(X_train , y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
MAE=mean_absolute_error(y_test , y_pred)
RMSE=np.sqrt(mean_squared_error(y_test , y_pred))
MAE , RMSE
```

# sklearn --> cross_validate

```python
X=df.drop('sales' , axis=1)
y=df['sales']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)

X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
model=Ridge(alpha=100)
```

```python
from sklearn.model_selection import cross_validate
```

```python
scores=cross_validate(model , X_train , y_train , cv=10 , scoring=['neg_mean_absolute_error' , 'neg_root_mean_squared_error'])
scores
```

```python
scores=pd.DataFrame(scores)
scores
```

```python
scores.mean()
```

```python
model=Ridge(alpha=1)
```

```python
score=cross_validate(model , X_train , y_train , scoring=['neg_mean_absolute_error' , 'neg_root_mean_squared_error'] , cv=10)
score=pd.DataFrame(score)
score
```

```python
score.mean()
```

```python
model.fit(X_train , y_train)
```

```python
y_pred= model.predict(X_test)
```

```python
MAE=mean_absolute_error(y_test , y_pred)
RMSE = np.sqrt(mean_squared_error(y_test , y_pred))
MAE , RMSE
```
