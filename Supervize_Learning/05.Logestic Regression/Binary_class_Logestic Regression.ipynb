## train test split

```python
X=df.drop('test_result' , axis=1)
y=df['test_result']

from sklearn.model_selection import train_test_split
X_train  , X_test , y_train , y_test=train_test_split(X , y , random_state=101 , test_size=0.1)


from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
X_train=scale.fit_transform(X_train)
X_test=scale.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression()
```

```python
log_model.fit(X_train  , y_train)
```

```python
log_model.predict(X_test)
```

```python
log_model.predict_proba(X_test)
```

```python
log_model.coef_
```

# Evaluating of model

```python
from sklearn.metrics import accuracy_score , recall_score , precision_score, classification_report,\
confusion_matrix  , ConfusionMatrixDisplay
```

```python
y_pred=log_model.predict(X_test)
```

```python
accuracy_score(y_test , y_pred)
```

```python
recall_score(y_test , y_pred)
```

```python
precision_score(y_test , y_pred)
```

# classification_report include all -------->precision , recall , f1-score

```python
print(classification_report(y_test , y_pred))
```

```python
confusion_matrix(y_test , y_pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test , y_pred)
```

```python
ConfusionMatrixDisplay.from_estimator(log_model , X_test , y_test)
```

```python
ConfusionMatrixDisplay.from_estimator(log_model , X_test , y_test , normalize='true')
```
