```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('StudentsPerformance.csv')
df
```

# joinplot()نمایش نمودار پراکندگی و توزیع در کنار هم

```python
sns.jointplot(data=df , x='math score' , y='reading score')
```

## kind

```python
sns.jointplot(data=df , x ='math score' , y = 'reading score' , kind='hex')# kind--> scatter , hist , kde , hex
sns.set()
```

# pairplot()

```python
sns.pairplot(data=df)
```

### corner

```python
sns.pairplot(data=df , hue='gender' , palette='viridis'  , corner=True)
```

## diag_kind

```python
sns.pairplot(data=df , hue ='gender' , palette='viridis' , corner=True , diag_kind='hist')
```

# catplot()

```python
sns.catplot(data=df , x='gender' , y='math score', hue='gender' , kind='box' , col='lunch')
```

```python
sns.catplot(data=df , x='gender' , y='math score' , hue='gender' , row='lunch' , col='test preparation course' , kind='box')
sns.set()
```

# PairGrid()

```python
g=sns.PairGrid(df , hue='gender')
g=g.map_upper(sns.scatterplot)
g=g.map_diag(sns.kdeplot)
g=g.map_lower(sns.kdeplot , color='r')
g=g.add_legend()
```

# FacetGrid()

```python
g=sns.FacetGrid(data=df, hue='gender' , col='gender' , row='lunch')
g=g.map(plt.scatter, 'math score' , 'reading score', edgecolor='w' )
g.add_legend(bbox_to_anchor=(1.1,0.25),  borderaxespad=0)
plt.subplots_adjust(hspace=0.4 , wspace=1)
```
