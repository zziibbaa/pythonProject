# کاهش حجم تصاویر

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

# خواندن تصویر

```python
img_as_array=mpimg.imread('palm_trees.jpg')
img_as_array
```

```python
plt.imshow(img_as_array)
```

```python
img_as_array.shape
```

```python
h , w , c = img_as_array.shape
print(h , w , c)
```

```python
img_as_array2d=img_as_array.reshape(h*w , c)
img_as_array2d.shape
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=7)
labels=model.fit_predict(img_as_array2d)
labels
```

```python
model.cluster_centers_
```

```python
rgb_code=model.cluster_centers_.round(0).astype(int)
rgb_code
```

```python
rgb_code[labels]
```

```python
rgb_code[labels].shape
```

```python
rgb_code.shape
```

```python
len(rgb_code[labels])
```

```python
quantize_img=rgb_code[labels].reshape(h , w , c)
quantize_img.shape
```

```python
quantize_img
```

```python
fig , axes=plt.subplots(ncols=2 , nrows=1)
axes[0].imshow(quantize_img)
axes[1].imshow(img_as_array)
plt.subplots_adjust(wspace=.3)
```
