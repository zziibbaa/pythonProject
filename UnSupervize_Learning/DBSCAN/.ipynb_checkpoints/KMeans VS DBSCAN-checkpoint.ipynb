```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
blobs=pd.read_csv('cluster_blobs.csv')
blobs.head()
```

```python
sns.scatterplot(data=blobs , x='X1' , y='X2')
```

```python
circles=pd.read_csv('cluster_circles.csv')
circles.head()
```

```python
sns.scatterplot(data=circles , x='X1' , y='X2')
```

```python
mones=pd.read_csv('cluster_moons.csv')
mones.head()
```

```python
sns.scatterplot(data=mones , x='X1' , y='X2')
```

```python
def display_model(model , data):
    labels=model.fit_predict(data)
    sns.scatterplot(data=data , x='X1' , y='X2' , hue=labels , palette='Set1')
```

```python
from sklearn.cluster import KMeans , DBSCAN
```

```python
kmean_model=KMeans(n_clusters=3)
display_model(kmean_model , blobs)
```

```python
DB_model=DBSCAN()
display_model(DB_model , blobs)#red sample is out_lyer know in DBSCAN
```

```python
DB_model=DBSCAN(eps=5)
display_model(DB_model , blobs)
```

```python
kmean_model=KMeans(n_clusters=2)
display_model(kmean_model , circles)
```

```python
DB_model=DBSCAN(eps=0.15)
display_model(DB_model , circles)
```

```python
kmean_model=KMeans(n_clusters=2)
display_model(kmean_model , mones)
```

```python
DB_model=DBSCAN(eps=0.15)
display_model(DB_model , mones)
```
