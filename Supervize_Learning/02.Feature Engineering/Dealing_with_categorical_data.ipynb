```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df2= pd.DataFrame({'country' : ['UK' , 'Germany' , 'Iran', 'Japan', 'Iran', 'Germany' , 'Japan']})


ohe=OneHotEncoder(sparse_output=False , dtype=np.int64)#drop='first'
df_encoded=ohe.fit_transform(df2[['country']])
df_encoded
```

```python
df_encoded= pd.DataFrame(df_encoded , columns=['Germany' , 'Iran' , 'Japan' , 'UK'])
df_encoded
```
