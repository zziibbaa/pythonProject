```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('Sales_Funnel_CRM.csv')
df
```

```python
licenses= df[['Company' , 'Product' , 'Licenses']]
licenses
```

```python
pd.pivot(data=licenses , index='Company' , columns='Product' , values='Licenses')
```

# pivot table for aggregate functions

```python
pd.pivot_table(df , index='Company' , aggfunc= 'sum')
```

```python
pd.pivot_table(df , index='Company' , aggfunc='sum')[['Licenses' , 'Sale Price']]
#pd.pivot_table(df , index='Company' , aggfunc='sum' , values= ['Licenses' , 'Sale Price']) another way
```

```python
#another way
df.groupby('Company').sum()[['Licenses' , 'Sale Price']]
```

```python
pd.pivot_table(df , index=['Account Manager' , 'Contact'] , values='Sale Price' , aggfunc='sum')
```

```python
pd.pivot_table(df , index=[ 'Account Manager' , 'Contact' ] , columns='Product' , values='Sale Price' ,
               aggfunc='sum' , fill_value=0)
```

```python
pd.pivot_table(df , index=[ 'Account Manager' , 'Contact' , 'Product']  , values='Sale Price' ,aggfunc=[np.sum , np.mean]
               , fill_value=0)
```

```python
#get final 'all' with margin = True
pd.pivot_table(df , index=[ 'Account Manager' , 'Contact' , 'Product']  , values='Sale Price' ,aggfunc=[np.sum , np.mean]
               , fill_value=0 , margins=True)
```

```python
pd.pivot_table(df , index=['Account Manager' , 'Contact' , 'Status'] , values='Sale Price' , aggfunc='sum' )
```
