```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.arange(0,10)
x
```

```python
y= 2*x
y
```

# plt.plot

```python
plt.plot(x,y)
plt.xlabel('x axis title here')
plt.ylabel('y axis title here')
plt.title('string title here' , loc='left')
```

# title , xlabel , ylabel , xlim , ylim

```python
plt.plot(x,y)
plt.xlabel('x axis title here')
plt.ylabel('y axis title here')
plt.title('string title here' , loc='left')

plt.xlim(0,6)
plt.ylim(0,12)
```

# savefig , dpi

```python
plt.plot(x,y)
plt.xlabel('x axis title here')
plt.ylabel('y axis title here')
plt.title('string title here' , loc='left')
plt.xlim(0,10)
plt.ylim(0,20)
plt.savefig('example.png' , dpi=300)
```

```python
a=np.linspace(0,10,11)
a
```

```python
b=a**4
b
```

```python
x=np.arange(0,10)
x
```

```python
y=x*2
y
```

# plt.figure , add_axes

```python
fig= plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes1.plot(x,y)
#plt.show()
```

```python
fig= plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.2,0.5,0.5])
axes1.plot(a,b)
axes2.plot(x,y)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')
axes2.set_title('Small Figure')
plt.show()
```

# zoom in 

```python
fig= plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,0.25,0.25])
axes1.plot(a,b)
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')
axes2.set_title('Zoom In')
plt.show()
```

#   multiplot

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.5,0.5,0.25,0.25])
axes3=fig.add_axes([1.1,0.5,0.5,0.5])
axes1.plot(a,b)
axes2.plot(a,b)
axes3.plot(x,y)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
```

# figsize , savefig

```python
fig=plt.figure(figsize=(8,6) , dpi=100)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
fig.savefig('fig.png' , bbox_inches='tight')#ذخیره سازی بصورت فشرده
```

# rcParams

```python
fig=plt.figure()
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 20
axes1=fig.add_axes([0,0,1,1])
axes1.plot(x,y)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
```

# legend()

```python
x = np.arange(0,10)
fig= plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x , x**2, label='x**2')
axes.plot(x ,x**3 , label='x**3')
axes.legend()#loc='upper right','upper left',0(best),2,...
```

# color & linestyle & alpha(0-1)-->transparency

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x , x+1 , 'b--')
axes.plot(x , x+2 , 'g:')
axes.plot(x , x+3 , color ='gray')
axes.plot(x , x+4 , 'y.-' )
axes.plot(x , x+5 , 'r-*')
axes.plot(x , x+6 , color='#00bb5d' , alpha=0.5)
```

# linewidth | lw

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x , x+1 , 'b--' , linewidth=0.25)
axes.plot(x , x+2 , 'b--' , lw=0.5)
axes.plot(x , x+3 , 'b--' , lw = 1)
axes.plot(x , x+4 , 'b--' , lw = 10)
```

# ax[0].set_

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
ax=axes.plot(x , x+1 , 'b--' , lw=5)
ax[0].set_dashes([10,10])
```

# marker & markersize & grid()

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x , x+1 ,  marker='o' , markersize=20)
axes.plot(x , x+2, marker='s' , ms=20)
axes.plot(x , x+3 , marker='3' , ms=20 , lw=0)
axes.plot(x , x+4 , marker='v' , ms = 20)
axes.plot(x , x+5 , marker='h' , ms=20)
plt.grid(True , ls='--' , lw=0.5 , alpha=0.7)
```
