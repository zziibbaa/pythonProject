```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('cancer_tumor_data_features.csv')
df.head()
```

# Standardazation of data

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
scaled_X= scale.fit_transform(df)
```

# Implementation PCA with scikit learn

```python
from sklearn.decomposition import PCA
```

```python
pca= PCA(n_components=2)
```

```python
principal_components=pca.fit_transform(scaled_X)
```

# show data with principal components

```python
plt.scatter(principal_components[: , 0] , principal_components[: , 1])
plt.xlabel('1st principal component')
plt.ylabel('2en principal component')
```

# show data with principal components & target labels

```python
from sklearn.datasets import load_breast_cancer
cancer_dict=load_breast_cancer()
cancer_dict.keys()
```

```python
cancer_dict['target']
```

```python
plt.figure(figsize=(12,6))
plt.scatter(principal_components[: , 0] , principal_components[: , 1] , c=cancer_dict['target'])
plt.xlabel('1st principal component')
plt.ylabel('2en principal component')
```

```python
pca.components_
```

```python
df_components=pd.DataFrame(pca.components_ , index=['pc1' , 'pc2'] , columns=df.columns)
df_components
```

```python
plt.figure(figsize=(20,3))
sns.heatmap(df_components , annot=True)
```

# calculate best value of number of components with ellbo method

```python
explaine_varience= []
for n in range(1,30):
    pca=PCA(n_components=n)
    pca.fit(scaled_X)
    explaine_varience.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,30), explaine_varience)
plt.xlabel('number of components')
plt.ylabel('varience explained')
```
