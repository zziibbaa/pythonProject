```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
m= np.linspace(0,10,11)
m
```

```python
c=3 * (10**8)
E=m* (c**2)
E
```

```python
plt.plot(m,E , 'r' , lw=5)
plt.xlabel('Mass in Grams')
plt.ylabel('Energy in Joules')
plt.title('E=mc^2')
plt.xlim(0,11)
#plt.ylim(0, 9.0e+17)
```
