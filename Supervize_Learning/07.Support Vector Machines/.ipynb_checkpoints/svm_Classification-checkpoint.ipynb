```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('mouse_viral_study.csv')
df.head()
```

```python
df.corr()
```

```python
sns.scatterplot(data=df ,x='Med_1_mL' , y='Med_2_mL' , hue='Virus Present')
x=np.linspace(1,10,100)
y=-x+11
plt.plot(x, y,'black')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from svm_margin_plot import plot_svm_boundary
from sklearn.model_selection import GridSearchCV
```

# هر چه مقدار c بزرگتر باشد نواحی مرزی سخت تر است و داده های کمتری در ناحیه ی مرزی قرار میگیرند اما هرچه کوچکتر باشد نواحی مرزی تعداد بیشتری داده قرار می گیرد

# kernel ----> linear , rbf , sigmoid , poly , precomputed

```python
X= df.drop('Virus Present' ,axis=1)
y=df['Virus Present']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
scale=StandardScaler()
X_train=scale.fit_transform(X_train)
X_test=scale.transform(X_test)
model=SVC(kernel='linear' , C=1000)
model.fit(X , y)
```

```python
plot_svm_boundary(model , X , y)
```

```python
model= SVC(kernel='linear' , C=.001)
model.fit(X , y)
plot_svm_boundary(model , X , y)
```

# نکته: هر چه گاما بزرگ تر باشد باعث اورفیت شدن مدل گاما کوچک تر بایاس مدل را بیشتر می کند  

```python
model=SVC(kernel='rbf' , C=1  , gamma='scale')
model.fit(X , y)
plot_svm_boundary(model , X , y)
```

```python
model= SVC(kernel='rbf' , C=1 , gamma='auto')
model.fit(X , y)
plot_svm_boundary(model , X , y)
```

```python
model=SVC(kernel='rbf' , C=1 , gamma=0.01)
model.fit(X , y)
plot_svm_boundary(model , X , y)
```

```python
model=SVC(kernel='rbf' , C=1, gamma=1)
model.fit(X , y)
plot_svm_boundary(model , X ,y)
```

```python
model=SVC(kernel='sigmoid' , C=1)
model.fit(X ,y)
plot_svm_boundary(model , X , y)
```

```python
model=SVC(kernel='poly' , degree=2 , C=1)
model.fit(X ,y)
plot_svm_boundary(model , X ,y)
```

# GridsearchCV

```python
svm=SVC()
param_grid={'C' : [.0001 , .001 , .01 , .1 , 1 , 10 , 100] , 'kernel' : ['linear' , 'rbf' , 'sigmoid' , 'poly'] , 'degree' : [1 ,2 , 3],
            'gamma' : ['scale','auto'] }
grid_model=GridSearchCV(svm , param_grid , cv=10)
grid_model.fit(X_train ,y_train)
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```

```python
grid_pred=grid_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report , ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_predictions(y_test , grid_pred)
```

```python
print(classification_report(y_test , grid_pred))
```
