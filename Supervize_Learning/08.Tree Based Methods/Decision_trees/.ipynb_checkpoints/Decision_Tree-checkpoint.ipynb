## همچنین نیاز به مقیاس گذاری داده ها نیست چون در هر مرحله یک ویژگی برای گره انتخاب می شود و تمام ویژگی ها با هم بررسی نمی شوند که مشکل تفاوت مقیاس ها را داشته باشیم

# train | test split

```python
X= pd.get_dummies(df.drop('species' , axis=1) , drop_first=True)
y=df['species']

from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.3 , random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
model.fit(X_train , y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report , ConfusionMatrixDisplay
```

```python
print(classification_report(y_test , y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test , y_pred)
```

```python
model.feature_importances_
```

```python
X.columns
```

```python
feature_important=pd.DataFrame(index=X.columns , data=model.feature_importances_ , columns=['Feature Importances'])
feature_important
```

```python
sns.boxplot(data=df , x='species' , y='body_mass_g' , hue='species')
#چون از لحاظ وزنی دو گونه شباهت بسیار زیادی بهم دارند ،معیار وزن چندان اثری در جداسازی گونه ها از هم ندارد
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(13,10))
plot_tree(model , filled=True , feature_names=X.columns);
```

```python
def report_paramiter(model , X_test , y_test):
    y_pred=model.predict(X_test)
    print(classification_report(y_test , y_pred))
    print('\n')
    plt.figure(figsize=(13,10))
    plot_tree(model , filled=True , feature_names=X_test.columns);
```

# max_depth حداکثر عمق

```python
prund_tree=DecisionTreeClassifier(max_depth=2)
prund_tree.fit(X_train , y_train)
report_paramiter(prund_tree , X_test , y_test)
```

# max_leaf_node حداکثر گره های برگ

```python
prund_tree=DecisionTreeClassifier(max_leaf_nodes=3)
prund_tree.fit(X_train , y_train)
report_paramiter(prund_tree , X_test , y_test)
```

# criterion --> entropy

```python
entropy_tree=DecisionTreeClassifier(criterion='entropy')
entropy_tree.fit(X_train , y_train)
report_paramiter(entropy_tree , X_test , y_test)
```
