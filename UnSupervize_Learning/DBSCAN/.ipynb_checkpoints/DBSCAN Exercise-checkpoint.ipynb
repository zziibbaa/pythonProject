```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
scale_df=pd.DataFrame(scale.fit_transform(df))
```

```python
from sklearn.cluster import DBSCAN
number_outleirs=[]
perc_outleirs=[]
for eps in np.linspace(0.001 , 10 , 100):
    db_model=DBSCAN(eps=eps , min_samples=16)
    db_model.fit(scale_df)
    number_outleirs.append(np.sum(db_model.labels_==-1))
    perc_outleirs.append(100*(np.sum(db_model.labels_==-1)/len(db_model.labels_)))
```

```python
sns.lineplot(x=np.linspace(0.001,10,100) , y=number_outleirs)
plt.xlabel('epsilon values')
plt.ylabel('number of points classifide outleirs')
```

```python
sns.lineplot(x=np.linspace(0.001, 10 , 100) , y=perc_outleirs)
plt.xlabel('epsilon values')
plt.ylabel('percentage of points classified outliers')
```

```python
sns.lineplot(x=np.linspace(0.001, 10 , 100) , y=perc_outleirs)
plt.xlabel('epsilon values')
plt.ylabel('percentage of points classified outliers')
plt.ylim(0,20)
plt.xlim(0,8)
plt.hlines(y=5 , xmin=0 , xmax=10 , colors='red', ls='--')
```

```python
model=DBSCAN(eps=2 , min_samples=16)
model.fit(scale_df)
np.unique(model.labels_)
```

```python
sns.scatterplot(data=df , x='Grocery' , y='Milk' , hue=model.labels_ , palette='Set1')
```

```python
sns.scatterplot(data=df , x='Milk' , y='Detergents_Paper' , hue= model.labels_ , palette='Set1')
```

```python
df['labels']=model.labels_
df.head()
```

```python
numerical_features=df.drop(['Channel' , 'Region'] , axis=1)
numerical_mean=numerical_features.groupby('labels').mean()
numerical_mean
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaled_df=MinMaxScaler()
data=scaled_df.fit_transform(numerical_mean)
numerical_mean=pd.DataFrame(data=data , columns=numerical_mean.columns , index=numerical_mean.index)
numerical_mean
```

```python
sns.heatmap(numerical_mean, annot=True)
```
