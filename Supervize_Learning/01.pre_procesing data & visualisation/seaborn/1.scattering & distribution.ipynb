```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('dm_office_sales.csv')
df.info()
```

```python
sns.scatterplot(data=df , x= 'salary' , y='sales')
```

# hue --> nominal or ordinal data recived

```python
plt.figure(figsize=(10,6) ,dpi=150)
sns.scatterplot(data=df , x='salary' , y='sales' , hue='division')
```

```python
plt.figure(figsize=(6,4),dpi=150)
sns.scatterplot(data=df , x='salary' , y='sales' , hue='work experience')
```

# pallete

```python
plt.figure(figsize=(6,4) , dpi=150)
sns.scatterplot(data=df , x= 'salary' , y= 'sales', hue='division' , palette='viridis')
```

# size --> ordinal data recevied

```python
plt.figure(figsize=(6,4) , dpi=150)
sns.scatterplot(data=df , x ='salary' , y='sales' , size='work experience')
```

## s & linewidth & alpha

```python
plt.figure(figsize=(6,4) , dpi=150)
sns.scatterplot(data = df , x='salary' , y='sales' , s=50 , linewidth=0 , alpha=0.5)
```

# style --> nominal or ordinal data recived & savefig

```python
plt.figure(figsize=(10,8) , dpi=200)
sns.scatterplot(data=df , x='salary' , y='sales' , style='level of education' , hue='level of education' , s=100)
plt.savefig('example of scatter.jpg')
```

# sns.set

```python
sns.set(font_scale=1.5 , style='whitegrid' , font='Times New Roman')
```

# rugplot

```python
sns.rugplot(data=df , x='salary')
```

```python
#height
sns.rugplot(data=df , x='salary' , height=0.2)
```

# displot

```python
#kde(kernel dencity estimation)
sns.displot(data=df , x='salary' , kde=True)
```

```python
#bins
sns.displot(data=df , x='salary' , bins=10)
```

```python
sns.set()
sns.displot(data=df , x='salary' , bins=100)
```

# histogram

## color & edgecolor & linewidth & linestyle

```python
sns.histplot(data=df , x='salary' , bins=20 , color='orange' , edgecolor='blue' , lw=3 , ls='--')
```

```python
np.random.seed(42)
sample_ages=np.random.randint(0,100,200)
```

```python
sample_ages=pd.DataFrame(sample_ages , columns=['ages'])
```

```python
sns.displot(data=sample_ages , x='ages' , bins=10 , rug=True , kde=True)
```

# kdeplot

```python
#clip --> cutter
sns.kdeplot(data=sample_ages , x='ages' , clip=[0,100], color='r' )
```

```python
#bw_adjust --> band width_adjust by defult=1
sns.kdeplot(data=sample_ages , x='ages' , bw_adjust=0.5 , fill=True , color='g')
```

# 2D kdeplot

```python
random_data=pd.DataFrame(np.random.normal(0,1 , size=(100,2)) , columns=['x' , 'y'])
sns.kdeplot(data=random_data , x='x' , y='y')
```
