```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# EDA

```python
df= pd.read_csv('data_banknote_authentication.csv')
df.info()
```

```python
sns.heatmap(df.corr() , cmap='coolwarm' , annot=True)
```

```python
sns.pairplot(df , hue='Class')
```

```python
sns.countplot(x='Class' , data=df , hue='Class')
```

```python
X=df.drop('Class' , axis=1)
y=df['Class']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
model.get_params().keys()
```

# bootstrap      1372              

### out_of_bag_error(OOB)         bootstrap                  

```python
n_estimators = [7, 64 , 100 , 128 , 170]
max_features = [2 , 3 , 4]
bootstrap = [ True , False]
oob_score = [True , False]

param_grid = {'n_estimators' : n_estimators , 
              'max_features' : max_features , 
              'bootstrap' : bootstrap , 
              'oob_score' : oob_score}

from sklearn.model_selection import GridSearchCV
grid_model = GridSearchCV(model , param_grid )
grid_model.fit(X_train , y_train)
```

```python
grid_model.best_params_
```

```python
y_pred= grid_model.predict(X_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay , classification_report , accuracy_score
```

```python
ConfusionMatrixDisplay.from_predictions(y_test , y_pred )
```

```python
print(classification_report(y_test , y_pred))
```

```python
error=[]
mis_clasification=[]
for n in range (1 , 64 ):
    model=RandomForestClassifier(n_estimators=n , max_features=2, bootstrap=True )
    model.fit(X_train , y_train)
    y_pred=model.predict(X_test)
    error.append(1-accuracy_score(y_test , y_pred))
    mis_clasification.append(np.sum(y_test != y_pred))
```

```python
error
```

```python
mis_clasification
```

```python
fig ,axis = plt.subplots(nrows=2 , ncols=1)
axis[0].plot(range(1,64) , error , 'r')
axis[1].plot(range(1,64) , mis_clasification , 'g')
plt.grid()
```
