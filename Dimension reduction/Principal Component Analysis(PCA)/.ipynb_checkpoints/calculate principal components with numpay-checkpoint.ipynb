```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('cancer_tumor_data_features.csv')
df.info()
```

# Standardazation of data

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
scaled_X=scale.fit_transform(df)
scaled_X[0]
```

# calculate covarians matrix

```python
covariance_matrix=np.cov(scaled_X , rowvar=False)
covariance_matrix
```

# calculate eigen vector & eigen value

```python
eigen_values , eigen_vectors = np.linalg.eig(covariance_matrix)
```

# principal components with value 2

```python
num_components=2
```

```python
# get index sorting key based on eigen values
sorted_keys=np.argsort(eigen_values)[::-1][:num_components]
```

```python
# get num_components of eigen values and eigen vectors
eigen_values , eigen_vectors = eigen_values[sorted_keys] , eigen_vectors[:,sorted_keys]
```

```python
principal_component=np.dot(scaled_X , eigen_vectors)
```

# show data with principal components

```python
plt.scatter(x=principal_component[: , 0] , y=principal_component[ : , 1])
plt.xlabel('1st principal component')
plt.ylabel('2en principal component')
```
