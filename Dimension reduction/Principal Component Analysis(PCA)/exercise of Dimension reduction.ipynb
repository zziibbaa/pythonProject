```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
X=scale.fit_transform(X)
X
```

```python
pca=PCA(n_components=2)
pca_pixels=pca.fit_transform(X)
pca_pixels
```

```python
pca.explained_variance_ratio_
```

```python
np.sum(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=pca_pixels[ : , 0] , y=pca_pixels[ : , 1] , hue=df['number_label'] , palette='turbo')
plt.legend(loc=(1.01 , 0.5))
plt.xlabel('1st principal components')
plt.ylabel('2ed principal components')
```

```python
explaiend_varience=[]
for n in range(1,64):
    pca_model=PCA(n_components=n)
    pca_model.fit(X)
    explaiend_varience.append(np.sum(pca_model.explained_variance_ratio_))
```

```python
plt.plot(range(1,64) , explaiend_varience)
plt.xlabel('number of component')
plt.ylabel('expained varience')
```

```python
pca=PCA(n_components=20)
pca.fit_transform(X)
pca.explained_variance_ratio_
```

```python
np.sum(pca.explained_variance_ratio_)
```

```python
import plotly.express as px
```

```python
pca=PCA(n_components=3)
pca_pixels=pca.fit_transform(X)
pixel=pd.DataFrame(pca_pixels , columns=['F1' , 'F2' , 'F3'])
pixel['label']=df['number_label']
```

```python
fig=px.scatter_3d(pixel ,
                  x='F1',
                  y='F2',
                  z='F3',
                  color='label',
                  title='3D PCA',
                  labels={'F1':'1st principal component',
                          'F2':'2nd principal component',
                          'F3':'3rd principal component'},
                  opacity=0.7,
                  color_discrete_sequence=px.colors.qualitative.Set2)

fig.update_layout(scene=dict(xaxis_title='1st principal component',
                             yaxis_title='2nd principal component',
                             zaxis_title='3rd principal component'),
                  width=800,
                  height=800,
                  showlegend=True)

fig.show()
```
