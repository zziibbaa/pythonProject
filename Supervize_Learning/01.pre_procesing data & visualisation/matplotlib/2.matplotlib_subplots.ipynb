```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a=np.linspace(0,10,11)
a
```

```python
b=a**4
b
```

```python
x=np.arange(0,10)
x
```

```python
y=2*x
y
```

# subplots()

```python
fig , axes=plt.subplots()
axes.plot(x,y , 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig , axes=plt.subplots(nrows=1 , ncols=2)
axes[0].plot(a,b)
axes[1].plot(x,y , 'r')
```

# fig.suptitle() &fig.subplots_adjust() && nrows & ncols 

```python
fig , axes=plt.subplots(nrows=2, ncols=2 , figsize=(12,6))
axes[0][0].plot(x,y)
axes[0][1].plot(a,b ,'y')
axes[1][0].plot(y,x , 'r')
axes[1][1].plot(b,a ,'g')
axes[0][0].set_title('X & Y plot')
axes[0][0].set_xlabel('X')
fig.suptitle('Figur Level', fontsize=16)
fig.subplots_adjust(wspace=0.5 , hspace=0.3)
#fig.savefig()
```
