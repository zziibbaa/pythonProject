# filling missing value with interpolate

```python
airline_tix = {'first':100 , 'business' : np.nan , 'economy-plus' : 50 , 'economy': 30}
```

```python
ser= pd.Series(airline_tix)
ser
```

```python
ser.interpolate()
```

```python
df= pd.DataFrame(ser , columns= ['Price'])
df
```

```python
df=df.reset_index()
```

```python
df
```

```python
df.interpolate(method='quadratic')
```
