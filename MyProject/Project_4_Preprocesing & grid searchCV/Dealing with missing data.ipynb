# Identify Missing data

```python
def missing_values(df):
    percent_nan=100 *(df.isnull().sum() / len(df))
    percent_nan= percent_nan[percent_nan>0].sort_values()
    return percent_nan
```

```python
percent_nan=missing_values(df)
percent_nan
```

```python
sns.barplot(x=percent_nan.index , y=percent_nan , hue=percent_nan , palette='husl')
plt.xticks(rotation=90);
```

## Identify Missing data less dane 1%

```python
sns.barplot(x=percent_nan.index , y=percent_nan , hue=percent_nan , palette='husl')
plt.xticks(rotation=90);
plt.ylim(0,1)
```

```python
percent_nan[percent_nan<1]
```
